```python
## build corpus with bigram
logging.getLogger().setLevel(logging.CRITICAL)

corpus = litstudy.build_corpus(docs_sd, ngram_threshold=0.8)
```

```python
## compute word distribution

logging.getLogger().setLevel(logging.CRITICAL)
litstudy.compute_word_distribution(corpus).filter(like='_', axis=0).sort_index()
```

```python
## visualise word distribution
plt.figure(figsize=(20, 3))
litstudy.plot_word_distribution(corpus, limit=50, title="Top words", vertical=True, label_rotation=45);
```

```python
%%time
## train NMF topic modelling
num_topics = 15
topic_model = litstudy.train_nmf_model(corpus, num_topics, max_iter=250)
```

```python
%%time
## NMF: top 5 words per topic

for i in range(num_topics):
    print(f'Topic {i+1}:', topic_model.best_tokens_for_topic(i))
```

```python
%%time
## NMF: word cloud

plt.figure(figsize=(15, 5))
litstudy.plot_topic_clouds(topic_model, ncols=5);
```

```python
%%time
## NMF: visualise non-linear embedding 'landscape' plot in 2D plane

plt.figure(figsize=(20, 20))
litstudy.plot_embedding(corpus, topic_model);
```

```python
##=====AdvanceTopicModelling=====##
```

```python
%%time

## find 
topic_id_cyber = topic_model.best_topic_for_token('cybersecurity')

## Get documents (articles) fitting to identified topics
i = 1
for doc_id in topic_model.best_documents_for_topic(topic_id_cyber, limit=25):
    print(f'{i} | [doc:{doc_id}] {docs_sd[int(doc_id)].title}')
    i +=1
```

```python
%%time

## find 
topic_id_dtech = topic_model.best_topic_for_token('digital')   #('digital' + ' ' + 'technology')

## Get documents (articles) fitting to identified topics
i = 1
for doc_id in topic_model.best_documents_for_topic(topic_id_dtech, limit=200):
    #print(f'{i} | [doc:{doc_id}] {docs_sd[int(doc_id)].title}')
    print(f'{i} | [doc:{doc_id}] | {docs_sd[int(doc_id)].publication_year} | {docs_sd[int(doc_id)].authors[:1]} | {docs_sd[int(doc_id)].title}')
    i +=1
```

```python
%%time
## annotate search result docs_sd document set with a "dl_topic" tag 
## for document that strongly belong to identified topic (i.e., weight above a certain threshold).

## define two groups: documents that have the tag "dl_topic"
## print the publications over the years to see if interest in topic1, topic2 has increased or decreased over the years.

threshold = 0.2

## topic_id_cyber
dl_topic_cyber = topic_model.doc2topic[:, topic_id_cyber] > threshold
#dl_topic = topic_model.doc2topic[:, topic_id_cyber] > threshold
dl_topic_dtech = topic_model.doc2topic[:, topic_id_dtech] > threshold
## #dl_topic = topic_model.doc2topic[:, topic_id] > threshold

#docs = docs.add_property('dl_topic', dl_topic)
#docs_sd = docs_sd.add_property('dl_topic', dl_topic_cyber)
docs_sd = docs_sd.add_property('dl_topic_cyber', dl_topic_cyber)
docs_sd = docs_sd.add_property('dl_topic_dtech', dl_topic_dtech)
#docs_sd = docs_sd.add_property('dl_topic_digital', dl_topic_dtech)

"""
groups = {
    'Digital Technology related': 'dl_topic_dtech',
    'other': 'not dl_topic',
}

"""
#'''
groups = {
    'Cybersecurity related': 'dl_topic_cyber',
    'Digital Technology related': 'dl_topic_dtech',
    'other': 'not dl_topic',
}
#'''

'''
groups = {
    'cybersecurity related': 'dl_topic_cyber',
    'Digital Technology related': 'dl_topic_dtech',
}
'''

## plot documents distribution per topic tag
litstudy.plot_year_histogram(docs_sd, groups=groups, stacked=True);
```

```python
## let us filter out year before 2018

#docs = docs_scopus.filter_docs(lambda d: d.publication_year >= 2000)
docs_sd_2010plus = docs_sd.filter_docs(lambda d: d.publication_year >= 2018)

## plot documents distribution per topic tag according to year filter
litstudy.plot_year_histogram(docs_sd_2010plus, groups=groups, stacked=True);
```
